```python
# Part two - Time for the ANN
# import the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout , BatchNormalization
from keras.utils import np_utils
from keras.optimizers import RMSprop, Adam
from keras import regularizers

```

```python
# Initialize the ANN
model = Sequential()

# Adding the input layer and the first hidden layer

# Not sure about what you want to do but if you want to add some L1/L2 regularization on your weights, you can use :
# keras.regularizers.l1(0.)
# keras.regularizers.l2(0.)
# keras.regularizers.l1_l2(0.)
# kernel_regularizer=regularizers.l2(0.01)
# kernel_regularizer=regularizers.l2(0.01)
model.add(Dense(64, input_dim=trainX.shape[1] , activation='relu', kernel_regularizer=regularizers.l1_l2(0.001)))
# model.add(Dense(64, input_dim=trainX.shape[1] , activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(Dense(196, activation='relu'))
model.add(Dense(32, activation='relu'))
# model.add(Dropout(0.5, seed=seed))
# model.add(Dropout(0.0001))
# model.add(Dropout())
model.add(Dense(10, activation='sigmoid'))

metrics_list = ['accuracy', 'mse', 'mae', 'mape', 'cosine']
# model.compile(optimizer = Adam(lr = 0.01),loss='sparse_categorical_crossentropy', metrics=metrics_list)
model.compile(optimizer = Adam(lr = 0.01, decay=0.01/50),loss='sparse_categorical_crossentropy', metrics=metrics_list)
```

```python
# Lets start to build the ANN model
# Initialize the ANN
# Part 4 - Evaluation, Improving and tunning the ANN

from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from keras.utils import np_utils
from keras.optimizers import RMSprop, Adam
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
def build_model():
    # Evaluate the ANN
    model = Sequential()
    model.add(Dense(64, input_dim=trainX.shape[1] , activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(128, activation='relu'))
    model.add(Dense(196, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(6, activation='sigmoid'))
    model.compile(optimizer = Adam(lr = 0.0005),loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
  
```

```python
model = KerasClassifier(build_fn = build_model, epochs=30 , batch_size = 256)
accuracies = cross_val_score(estimator = model, X=trainX, y=trainY, cv=5, n_jobs=-1)
```
